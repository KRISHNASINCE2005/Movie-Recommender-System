```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("C:\\Desktop\\MOVIE RECOMMENDER SYSTEM\\datasets\\tmdb_5000_movies.csv")
credits = pd.read_csv("C:\\Desktop\\MOVIE RECOMMENDER SYSTEM\\datasets\\tmdb_5000_credits.csv")
```

```python
movies = movies.merge(credits,on='title')
movies.head(5)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
vector = tfidf.fit_transform(df['tags']).toarray()
vector[0:5]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity[2]
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].title)
```

```python
import pickle
pickle.dump(df,open('models/movie_list.pkl','wb'))
pickle.dump(similarity,open('models/similarity.pkl','wb'))
```

```python
import ast

def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter += 1
    return ','.join(L)  # Join the list into a comma-separated string

# Apply the function to the 'cast' column
movies1['cast'] = movies1['cast'].apply(convert3)
```
